```python
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep

import requests
import os
```

```python
def download_images(driver):
    img_url_dic = []
    count = 0
    pos = 0
    
    for i in range(30):
        js = 'var q=document.documentElement.scrollTop=' + str(pos)
        driver.execute_script(js)
        pos += 500
        sleep(2)
    
        #find image
        img_elements = driver.find_elements("tag name","img")
        for img in img_elements:
            img_url = img.get_attribute('src')

            if isinstance(img_url,str):
                if 'images' in img_url and "gettyimages" not in img_url:
                    if img_url not in img_url_dic:
                        try:
                            print(img_url)
                            img_url_dic.append(img_url)
                            filename = "./Images/benedict/" + "benedict" + str(count) + ".jpg"
                            r = requests.get(img_url)
                            with open(filename,"wb") as f:
                                f.write(r.content)
                            f.close()
                            count += 1
                            print(str(count) + " image")
                            sleep(0.5)
                        except:
                            print('failure')
```

```python
keyword = "benedictcumberbatch"
url = "https://www.google.com/search?q="+keyword+"&tbm=isch"

path = './Images/benedict'
if not os.path.exists(path):
    os.makedirs(path)
```

```python
driver = webdriver.Chrome()
driver.get(url)

# click Cookie
driver.find_element("xpath","/html/body/c-wiz/div/div/div/div[2]/div[1]/div[3]/div[1]/div[1]/form[2]/div/div/button").click()

driver.maximize_window()
download_images(driver)
print('\nFinished!')
driver.quit()
```
